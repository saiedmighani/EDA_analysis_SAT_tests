# Merging the cleaned data

```python
#Imports:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_test_2017 = pd.read_csv("../a_data/tests_2017.csv")
df_test_2018 = pd.read_csv("../a_data/tests_2018.csv")

df_final = pd.merge(df_test_2017, df_test_2018, how='inner', on = "state")
df_final.info()
```
